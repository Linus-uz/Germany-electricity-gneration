```python
##  What I have noticed so far...

# Column names are too long, would be easier for analysis if there are simplified.
# The "End date" column is not necessary, the "Start date" column can just be named "Date" for simplicity.
# The "Nuclear [MWh] Calculated resolutions" column contains about 702 "-" charaters, which might affect future analysis.
# The data types are Objects, it would be nice to have the "Date" column as "Datetype", and the other columns as "Floats".

## I'd try to resolve these in the next steps....Let's gooo!!!!
```

```python
# Making the column names easily readable

Generation_data.columns = [col.replace(' [MWh] Calculated resolutions', '').replace(' ', '_') for col in Generation_data.columns]
Generation_data.head(1)
```

```python
# Renaming the "Start date" column to "Date"

Generation_data.rename(columns={'Start_date': 'Date'}, inplace=True)
```

```python
# Dropping the "End date" column

Generation_data.drop(columns=['End_date'], inplace=True)
```

```python
# Quick peek at the dataset after the changes

Generation_data.head(1)
```

```python
# Screening the Entire DataFrame for ('-') characters and replacing them with a "0"

check = Generation_data.isin(['-']) 

if check.any().any():
    # If any occurrences are found
    print("Unwanted characters ('-') found in the DataFrame.")
    
    # Print the rows where '-' occurs
    print("Occurrences of '-' in the DataFrame:")
    print(Generation_data[check.any(axis=1)])
    
    # Replace All Occurrences of '-' with 0 in the Entire DataFrame
    Generation_data.replace('-', 0, inplace=True)
    print("All occurrences of '-' have been replaced with 0.")
else:
    # If no occurrences are found
    print("No unwanted characters ('-') found in the DataFrame.")
```

```python
# Formatting the data column to date type

Generation_data['Date'] = pd.to_datetime(Generation_data['Date'])
```

```python
# Formatting the rest of the columns to float type

for col in Generation_data.columns:
    if col not in ["Date"]:
        # Remove commas before converting to float
        Generation_data[col] = Generation_data[col].astype(str).str.replace(",", "").astype("float32")
```

```python
# Did the changes work? Let's see the summary statistics again

Generation_data.describe()
```

```python
# Lets see the datatypes of the columns again

print(Generation_data.dtypes)
```

```python
# Saving the cleaned dataset to a new CSV file

Generation_data.to_csv("../data/processed/generation_cleaned.csv", index=False)
```
